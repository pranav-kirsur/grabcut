For this assignment you will implement the GrabCut method mentioned in this [paper](https://cvg.ethz.ch/teaching/cvl/2012/grabcut-siggraph04.pdf). It is essentially an iterative version of GraphCut as shown in the figure below. 


The code below takes an input image and follows these steps:
- It requires a bounding box to be drawn by the user to roughly segment out the foreground pixels
- It runs an initial min-cut optimization using the provided annotation
- The result of this optimization gives an initial segmentation 
- To further refine this segmentation, the user provides two kinds of strokes to aid the optimization
    - strokes on the background pixels
    - strokes on the foreground pixels
- The algorithm now utilizes this to refine the original segmentation

```python
# Constants for trimap
TRIMAP_BG = 0
TRIMAP_UNK = 1
TRIMAP_FG = 2

# Constants for grabcut
MATTE_BG = 0
MATTE_FG = 1
```

```python
#  beta = 1/(2*avg(||color[m] - color[n]||)**2)
def compute_beta(image):
    """Compute beta for the given image"""
    h, w = image.shape[0], image.shape[1]
    
    beta = 0
    
    # Iterate over the pixels and add up contributions from neigbours
    for i in range(h):
        for j in range(w):
            color = image[i,j]
            
            # Go through neighbours
            if i>0:
                # Directly vertical neighbour
                difference = image[i-1, j] - color
                dot_product = np.sum(difference * difference)
                beta+=dot_product
            if j>0:
                # Left neighbour
                difference = image[i, j - 1] - color
                dot_product = np.sum(difference * difference)
                beta+=dot_product
            if i>0 and j>0:
                # Up-left neigbour
                difference = image[i-1, j-1] - color
                dot_product = np.sum(difference * difference)
                beta+=dot_product
            if i>0 and j+1<w:
                # Up right neighbour
                difference = image[i-1, j+1] - color
                dot_product = np.sum(difference * difference)
                beta+=dot_product
                
    # Can be calculated by counting
    number_of_neighbour_pairs = 4*h*w + (-3*h) + (-3*w) + 2
    
    # Find beta
    beta = beta / number_of_neighbour_pairs
    beta = 1.0/(2.0 * beta)
    
    return beta 
```

```python
# Functions to make graph conversion easier
def coordinates_to_node(i, j, image):
    """Returns node in graph corresponding to coordinate, with top left of image being 1"""
    h, w = image.shape[0], image.shape[1]
    
    return 1 + i*w + j

def node_number_to_coordinates(node_number, image):
    """Returns coordinates in image corresponding to node number"""
    h, w = image.shape[0], image.shape[1]
    
    node_number -=1
    j = node_number % w
    i = int((node_number - j)/w)
    
    return i, j
```

```python
# Calculate n-weights for the graph
def calculate_n_weights(image, beta, gamma):
    """Returns a dictionary of the n-weights for the image with tuple of node numbers as keys"""
    h, w = image.shape[0], image.shape[1]
    
    n_weights = {}
    
    # Iterate over the pixels and calculate weights for the neighbours
    for i in range(h):
        for j in range(w):
            color = image[i,j]
            node_number = coordinates_to_node(i, j, image)
            
            # Go through neighbours
            if i>0:
                # Directly vertical neighbour
                difference = image[i-1, j] - color
                dot_product = np.dot(difference, difference)
                weight = gamma * np.exp(-beta * dot_product)
                
                # Find neigbour node number
                neighbour_node_number = coordinates_to_node(i - 1, j, image)
                
                # Add weight to dictionary
                n_weights[(node_number, neighbour_node_number)] = weight
                n_weights[(neighbour_node_number, node_number)] = weight
                
            
            if j>0:
                # Left neighbour
                difference = image[i, j - 1] - color
                dot_product = np.dot(difference, difference)
                weight = gamma * np.exp(-beta * dot_product)
                
                # Find neigbour node number
                neighbour_node_number = coordinates_to_node(i, j - 1, image)
                
                # Add weight to dictionary
                n_weights[(node_number, neighbour_node_number)] = weight
                n_weights[(neighbour_node_number, node_number)] = weight
                
                
            if i>0 and j>0:
                # Up-left neigbour
                difference = image[i-1, j-1] - color
                dot_product = np.dot(difference, difference)
                weight = gamma * np.exp(-beta * dot_product) * np.sqrt(0.5)
                
                # Find neigbour node number
                neighbour_node_number = coordinates_to_node(i - 1, j - 1, image)
                
                # Add weight to dictionary
                n_weights[(node_number, neighbour_node_number)] = weight
                n_weights[(neighbour_node_number, node_number)] = weight
                
                
            if i>0 and j+1<w:
                # Up right neighbour
                difference = image[i-1, j+1] - color
                dot_product = np.dot(difference, difference)
                weight = gamma * np.exp(-beta * dot_product) * np.sqrt(0.5)
                
                # Find neigbour node number
                neighbour_node_number = coordinates_to_node(i - 1, j + 1, image)
                
                # Add weight to dictionary
                n_weights[(node_number, neighbour_node_number)] = weight
                n_weights[(neighbour_node_number, node_number)] = weight
                
            
    return n_weights
```

```python
# Build graph for mincut
def build_graph(image, trimap, matte, fg_gmm, bg_gmm, n_weights, lamda, LARGE_CONSTANT=1e20):
    """Returns graph built from image"""
    image = image.copy()
    
    h, w = image.shape[0], image.shape[1]
    
    edge_list = []
    edge_capacities = []
    
    # We assume source is 0 and sink is hw + 1
    source = 0
    sink = h*w + 1
    
    # Get colors of the image
    image_colors = [np.ravel(image[:,:,0]), np.ravel(image[:,:,1]), np.ravel(image[:,:,2])]
    image_colors = np.transpose(image_colors)
    
    # Unknown indices
    unknown_indices = np.nonzero(trimap.reshape(-1) == TRIMAP_UNK)
    
    # Get log likelihoods for foreground and background
    fg_likelihoods = np.log(fg_gmm.calc_prob(image.reshape(-1,3)[unknown_indices]))
    bg_likelihoods = np.log(bg_gmm.calc_prob(image.reshape(-1,3)[unknown_indices]))
    
    likelihood_index = 0
    
    
    # Add edges to source and sink
    for i in range(h):
        for j in range(w):
            node_number = coordinates_to_node(i, j, image)
            
            if trimap[i,j] == TRIMAP_BG:
                # Add source edge
                edge_list.append([source, node_number])
                edge_capacities.append(0)
                
                # Add sink edge
                edge_list.append([node_number, sink])
                edge_capacities.append(LARGE_CONSTANT)

            elif trimap[i,j] == TRIMAP_FG:
                # Add sink edge
                edge_list.append([sink, node_number])
                edge_capacities.append(0)
                
                # Add source edge
                edge_list.append([node_number, source])
                edge_capacities.append(LARGE_CONSTANT)
                
            elif trimap[i,j] == TRIMAP_UNK:
                # Find d from the likelihoods
                d_fg = -fg_likelihoods[likelihood_index]
                d_bg = -bg_likelihoods[likelihood_index]
                
                likelihood_index+=1
                
                # Add source edge
                edge_list.append([source, node_number])
                edge_capacities.append(max(d_bg,0))
                
                
                # Add sink edge
                edge_list.append([node_number, sink])
                edge_capacities.append(max(d_fg,0))
                
                

    
    # Iterate over the pixels and add up contributions from neigbours
    for i in range(h):
        for j in range(w):
            
            node_number = coordinates_to_node(i, j, image)
            
            # Go through neighbours
            if i>0:
                # Directly vertical neighbour
    
                # Find neigbour node number
                neighbour_node_number = coordinates_to_node(i - 1, j, image)
            
                # Find weight of edge
                indicator = (matte[i,j] != matte[i - 1, j])
                indicator = 1
                weight = n_weights[(node_number, neighbour_node_number)] * indicator
                
                # Add to edge list and edge capacity list
                edge_list.append([node_number, neighbour_node_number])
                edge_capacities.append(weight)
                
            if j>0:
                # Left neighbour
                # Find neigbour node number
                neighbour_node_number = coordinates_to_node(i, j - 1, image)
                
                # Find weight of edge
                indicator = (matte[i,j] != matte[i, j - 1])
                indicator = 1
                weight = n_weights[(node_number, neighbour_node_number)] * indicator
                
                # Add to edge list and edge capacity list
                edge_list.append([node_number, neighbour_node_number])
                edge_capacities.append(weight)
                
            if i>0 and j>0:
                # Up-left neigbour
                # Find neigbour node number
                neighbour_node_number = coordinates_to_node(i - 1, j - 1, image)
                
                # Find weight of edge
                indicator = (matte[i,j] != matte[i - 1, j - 1])
                indicator = 1
                weight = n_weights[(node_number, neighbour_node_number)] * indicator
                
                # Add to edge list and edge capacity list
                edge_list.append([node_number, neighbour_node_number])
                edge_capacities.append(weight)                
                
            if i>0 and j+1<w:
                # Up right neighbour
                # Find neigbour node number
                neighbour_node_number = coordinates_to_node(i - 1, j + 1, image)
                
                # Find weight of edge
                indicator = (matte[i,j] != matte[i - 1, j + 1])
                indicator = 1
                weight = n_weights[(node_number, neighbour_node_number)] * indicator
                
                # Add to edge list and edge capacity list
                edge_list.append([node_number, neighbour_node_number])
                edge_capacities.append(weight)

                
            

    
    graph = igraph.Graph(n = h * w + 2, edges=edge_list, directed=False, edge_attrs={'capacity': edge_capacities})
    return graph
```

```python
# Runs the grabcut algorithm
def grabcut(image, trimap, k=5, num_iters=3, gamma=50, use_prev_matte=False, prev_matte=None, return_intermediate = False):
    """Runs grabcut algorithm using the provided trimap and returns a segmentation mask"""
    
    print('Started grabcut')
    # Height and width of image
    h, w = image.shape[0], image.shape[1]
    
    intermediate_results = []
    
    image = image.copy()
    image = image.astype('float32')
    
    # Set lambda
    lamda = 9 * gamma
    
    # Compute beta
    print("Computing beta...")
    beta = compute_beta(image)
    
    # Calculate n-weights(only need to be computed once)
    print("Computing n-weights...")
    n_weights = calculate_n_weights(image, beta, gamma)
    
    matte = trimap.copy()
        
    if use_prev_matte == False:
        # Initialise matte from trimap
        matte[trimap == TRIMAP_BG] = MATTE_BG
        matte[trimap == TRIMAP_UNK] = MATTE_FG
        matte[trimap == TRIMAP_FG] = MATTE_FG
    else:
        # Initialise matte from trimap and previous matte
        matte = prev_matte.copy()
        matte[trimap == TRIMAP_BG] = MATTE_BG
        matte[trimap == TRIMAP_FG] = MATTE_FG
    
    # Get fg and bg colors for GMMs according to matte
    fg_mask = matte == MATTE_FG
    bg_mask = matte == MATTE_BG
    
#     fg_colors = [np.ravel((image[:,:,0])[fg_mask]), np.ravel((image[:,:,1])[fg_mask]), np.ravel((image[:,:,2])[fg_mask])]
#     fg_colors = np.transpose(fg_colors)
    
#     bg_colors = [np.ravel((image[:,:,0])[bg_mask]), np.ravel((image[:,:,1])[bg_mask]), np.ravel((image[:,:,2])[bg_mask])]
#     bg_colors = np.transpose(bg_colors)
    
    if return_intermediate:
        img2 = image.astype('uint8')
        mask2 = np.where((matte == MATTE_FG), 255, 0).astype('uint8')
        output = cv2.bitwise_and(img2, img2, mask = mask2)
        intermediate_results.append(output)        
    
    # Get fg and bg indices
    fg_indices = np.nonzero(fg_mask)
    bg_indices = np.nonzero(bg_mask)
    
    # Get fg and bg colors from indices
    fg_colors = image[fg_indices]
    bg_colors = image[bg_indices]

        
    # Initialise foreground and background gmms using fg and bg
    fg_gmm = GaussianMixture(fg_colors, n_components=k)
    bg_gmm = GaussianMixture(bg_colors, n_components=k)
    

    for iteration in range(num_iters):
        print("Grabcut iteration", iteration)

        # Assign gmm components
        component_index = np.full((h,w), -1)
        component_index[fg_indices] = fg_gmm.which_component(fg_colors)
        component_index[bg_indices] = fg_gmm.which_component(bg_colors)
        
        # Fit gmms using component_index and fg, bg indices
        print("Fitting gmms")
        fg_gmm.fit(image[fg_indices], component_index[fg_indices])
        bg_gmm.fit(image[bg_indices], component_index[bg_indices])        

        # Build graph for graphcut
        print("Building graph...")
        graph = build_graph(image, trimap, matte, fg_gmm, bg_gmm, n_weights, lamda)
    
        # Find mincut of graph
        source = 0
        sink = h * w + 1
        mincut = graph.st_mincut(source, sink, capacity="capacity")
    
        # Get new mask generated by graphcut
        source_nodes = mincut.partition[0]

        mask = np.full(matte.shape, MATTE_BG)
        for node_number in source_nodes:
            if node_number!=source:
                node_coordinates = node_number_to_coordinates(node_number, image)
                mask[node_coordinates] = MATTE_FG
                
        
        # Generate new matte from trimap and mask
        matte = mask.copy()
        matte[trimap == TRIMAP_BG] = MATTE_BG
        matte[trimap == TRIMAP_FG] = MATTE_FG
        
#         # debug
        if return_intermediate:
            img2 = image.astype('uint8')
            mask2 = np.where((matte == MATTE_FG), 255, 0).astype('uint8')
            output = cv2.bitwise_and(img2, img2, mask = mask2)
            intermediate_results.append(output)

        
        
        # Set fg and bg indices for next iteration
        
        # Get fg and bg colors for GMMs according to matte
        fg_mask = matte == MATTE_FG
        bg_mask = matte == MATTE_BG

        # Get fg and bg indices
        fg_indices = np.nonzero(fg_mask)
        bg_indices = np.nonzero(bg_mask)
        
        # Get fg and bg colors from indices
        fg_colors = image[fg_indices]
        bg_colors = image[bg_indices]
        

    print('grabcut running finished')
    if return_intermediate:
        return matte, intermediate_results
    return matte
```

```python
class EventHandler:
    """
    Class for handling user input during segmentation iterations 
    """
    
    def __init__(self, flags, img, _mask, colors):
        
        self.FLAGS = flags
        self.ix = -1
        self.iy = -1
        self.img = img
        self.img2 = self.img.copy()
        self._mask = _mask
        self.COLORS = colors

    @property
    def image(self):
        return self.img
    
    @image.setter
    def image(self, img):
        self.img = img
        
    @property
    def mask(self):
        return self._mask
    
    @mask.setter
    def mask(self, _mask):
        self._mask = _mask
    
    @property
    def flags(self):
        return self.FLAGS 
    
    @flags.setter
    def flags(self, flags):
        self.FLAGS = flags
    
    def handler(self, event, x, y, flags, param):

        # Draw the rectangle first
        if event == cv2.EVENT_LBUTTONDOWN:
            if self.FLAGS['rect_over'] == False: 
                self.FLAGS['DRAW_RECT'] = True
                self.ix, self.iy = x,y

        elif event == cv2.EVENT_MOUSEMOVE:
            if self.FLAGS['DRAW_RECT'] == True:
                self.img = self.img2.copy()
                cv2.rectangle(self.img, (self.ix, self.iy), (x, y), self.COLORS['BLUE'], 2)
                self.FLAGS['RECT'] = (min(self.ix, x), min(self.iy, y), abs(self.ix - x), abs(self.iy - y))
                self.FLAGS['rect_or_mask'] = 0

        elif event == cv2.EVENT_LBUTTONUP:
            if self.FLAGS['rect_over'] == False: 
                self.FLAGS['DRAW_RECT'] = False
                self.FLAGS['rect_over'] = True
                cv2.rectangle(self.img, (self.ix, self.iy), (x, y), self.COLORS['BLUE'], 2)
                self.FLAGS['RECT'] = (min(self.ix, x), min(self.iy, y), abs(self.ix - x), abs(self.iy - y))
                self.FLAGS['rect_or_mask'] = 0

        
        # Draw strokes for refinement 

        if event == cv2.EVENT_LBUTTONDOWN:
            if self.FLAGS['rect_over'] == False:
                print('Draw the rectangle first.')
            else:
                self.FLAGS['DRAW_STROKE'] = True
                cv2.circle(self.img, (x,y), 3, self.FLAGS['value']['color'], -1)
                cv2.circle(self._mask, (x,y), 3, self.FLAGS['value']['val'], -1)

        elif event == cv2.EVENT_MOUSEMOVE:
            if self.FLAGS['DRAW_STROKE'] == True:
                cv2.circle(self.img, (x, y), 3, self.FLAGS['value']['color'], -1)
                cv2.circle(self._mask, (x, y), 3, self.FLAGS['value']['val'], -1)

        elif event == cv2.EVENT_LBUTTONUP:
            if self.FLAGS['DRAW_STROKE'] == True:
                self.FLAGS['DRAW_STROKE'] = False
                cv2.circle(self.img, (x, y), 3, self.FLAGS['value']['color'], -1)
                cv2.circle(self._mask, (x, y), 3, self.FLAGS['value']['val'], -1)
```

```python
def run(filename: str, rect_iters=3, user_edit_iters=2, return_images=False, return_intermediate=False,k=5,gamma=50):
    """
    Main loop that implements GrabCut. 
    
    Input
    -----
    filename (str) : Path to image
    """
    
    COLORS = {
    'BLACK' : [0,0,0],
    'RED'   : [0, 0, 255],
    'GREEN' : [0, 255, 0],
    'BLUE'  : [255, 0, 0],
    'WHITE' : [255,255,255]
    }

    DRAW_BG = {'color' : COLORS['BLACK'], 'val' : 0}
    DRAW_FG = {'color' : COLORS['WHITE'], 'val' : 1}

    FLAGS = {
        'RECT' : (0, 0, 1, 1),
        'DRAW_STROKE': False,         # flag for drawing strokes
        'DRAW_RECT' : False,          # flag for drawing rectangle
        'rect_over' : False,          # flag to check if rectangle is  drawn
        'rect_or_mask' : -1,          # flag for selecting rectangle or stroke mode
        'value' : DRAW_FG,            # drawing strokes initialized to mark foreground
        'grabcut_rectangle_phase': True,
    }

    img = cv2.imread(filename)
    img2 = img.copy()                                
    mask = np.full(img.shape[:2], 2, dtype = np.uint8) # mask is a binary array with : 0 - background pixels
                                                     #                               1 - foreground pixels 
    output = np.zeros(img.shape, np.uint8)           # output image to be shown
    grabcut_mask = np.zeros(img.shape[:2], np.uint8)
    
    # Input and segmentation windows
    cv2.namedWindow('Input Image')
    cv2.namedWindow('Segmented image')
    
    EventObj = EventHandler(FLAGS, img, mask, COLORS)
    cv2.setMouseCallback('Input Image', EventObj.handler)
    cv2.moveWindow('Input Image', img.shape[1] + 10, 90)

    while(1):
        
        img = EventObj.image
        mask = EventObj.mask
        FLAGS = EventObj.flags
        cv2.imshow('Segmented image', output)
        cv2.imshow('Input Image', img)
        
        k = cv2.waitKey(1)

        # key bindings
        if k == 27:
            # esc to exit
            break
        
        elif k == ord('0'): 
            # Strokes for background
            FLAGS['value'] = DRAW_BG
        
        elif k == ord('1'):
            # FG drawing
            FLAGS['value'] = DRAW_FG
        
        elif k == ord('r'):
            # reset everything
            FLAGS['RECT'] = (0, 0, 1, 1)
            FLAGS['DRAW_STROKE'] = False
            FLAGS['DRAW_RECT'] = False
            FLAGS['rect_or_mask'] = -1
            FLAGS['rect_over'] = False
            FLAGS['value'] = DRAW_FG
            FLAGS['grabcut_rectangle_phase'] = True
            img = img2.copy()
            mask = np.zeros(img.shape[:2], dtype = np.uint8) 
            EventObj.image = img
            EventObj.mask = mask
            output = np.zeros(img.shape, np.uint8)
        
        elif k == 13: 
            # Press carriage return to initiate segmentation
            
            #-------------------------------------------------#
            # Implement GrabCut here.                         #  
            # Function should return a mask which can be used #
            # to segment the original image as shown on L90   # 
            #-------------------------------------------------#
            print("enter pressed")
            
            # Rectangle drawing phase
            if FLAGS['grabcut_rectangle_phase']:
                print("here")
                
                # Generate trimap from user input(rectangle)
                trimap = np.full(img.shape[:2], TRIMAP_BG)
                rect_x, rect_y, rect_w, rect_h  = EventObj.FLAGS['RECT']
                trimap[rect_y:rect_y + rect_h, rect_x:rect_x + rect_w] = TRIMAP_UNK
                
                if return_intermediate:
                    grabcut_mask, intermed1 = grabcut(img2, trimap, num_iters=rect_iters, return_intermediate=True, k=k, gamma=gamma)
                else:
                    grabcut_mask = grabcut(img2, trimap, num_iters=rect_iters,k=k,gamma=gamma)

                EventObj.flags = FLAGS
                mask2 = np.where((grabcut_mask == 1), 255, 0).astype('uint8')
                output = cv2.bitwise_and(img2, img2, mask = mask2)
                
                FLAGS['grabcut_rectangle_phase'] = False

            
            # Refining phase using strokes
            else:
                # Generate trimap from rectangle as well as user strokes
                trimap = np.full(img.shape[:2], TRIMAP_BG)
                rect_x, rect_y, rect_w, rect_h  = EventObj.FLAGS['RECT']
                trimap[rect_y:rect_y + rect_h, rect_x:rect_x + rect_w] = TRIMAP_UNK
                
                # Add strokes to trimap
                trimap[mask == 1] = TRIMAP_FG
                trimap[mask == 0] = TRIMAP_BG
                
                
                # Refine grabcut segmentation
                grabcut_mask = grabcut(img2, trimap, use_prev_matte=True, prev_matte=grabcut_mask, num_iters=user_edit_iters,k=k,gamma=gamma)
                
                EventObj.flags = FLAGS
                mask2 = np.where((grabcut_mask == 1), 255, 0).astype('uint8')
                output = cv2.bitwise_and(img2, img2, mask=mask2)

    if return_intermediate:
        return img, intermed1, output
    if return_images:
        return img, output
                
```

# Gamma = 400

```python
img, output = run("../images/llama.jpg", return_images = True, rect_iters=5, gamma = 400)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(5,5))
plt.title("Gamma = 400")
output = cv2.cvtColor(output, cv2.COLOR_BGR2RGB)
plt.imshow(output)
```

Thus after trying gamma values of 5, 50 and 400 we see that 50 gives the best results.
This is because gamma is proportional to the n-weights of the graph.
It is weightage given to smoothness term.
As described in "Interactive Image Segmentation Using an Adaptive GMMRF Model" ( referenced in Grabcut paper), setting it too high has an effect of shrinking the image, while setting it too low tends to generate isolated segmented regions.

# Size of the bounding box

## Big bounding box

```python
img, output = run("../images/llama.jpg", return_images = True, rect_iters=3)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(10,10))
plt.title("Big bounding box")
output = cv2.cvtColor(output, cv2.COLOR_BGR2RGB)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(np.hstack((img,output)))
```

## Tight bounding box

```python
img, output = run("../images/llama.jpg", return_images = True, rect_iters=3)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(10,10))
plt.title("Tight bounding box")
output = cv2.cvtColor(output, cv2.COLOR_BGR2RGB)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(np.hstack((img,output)))
```

Thus, we see that a tight bounding box helps the algorithm achieve a much better segmentation, for a given number of iteration that the algorithm is supposed to be run for.
This is because the bounding box specifies a "confirmed background for the algorithm".

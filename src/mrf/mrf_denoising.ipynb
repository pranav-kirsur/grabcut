```python
import numpy as np
import cv2
import igraph
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("./yin_and_yang.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
image = image.astype('int')
```

```python
plt.imshow(image, cmap = 'gray')
```

```python
# Variance of gaussian
sigma = 50
```

```python
gaussian_noise = np.random.normal(0, sigma, (image.shape[0],image.shape[1]))
```

```python
# Adding gaussian noise
noisy_image = image + gaussian_noise
noisy_image = np.clip(noisy_image, 0, 255)
```

```python
plt.imshow(noisy_image, cmap='gray')
```

```python
# Functions to make graph conversion easier
def coordinates_to_node(i, j, image):
    """Returns node in graph corresponding to coordinate, with top left of image being 1"""
    h, w = image.shape[0], image.shape[1]
    
    return 1 + i*w + j

def node_number_to_coordinates(node_number, image):
    """Returns coordinates in image corresponding to node number"""
    h, w = image.shape[0], image.shape[1]
    
    node_number -=1
    j = node_number % w
    i = int((node_number - j)/w)
    
    return i, j
```

```python
# Build graph for denoising
def build_graph(image, regularisation_param):
    """Returns graph built from image"""
    image = image.copy()
    
    h, w = image.shape[0], image.shape[1]
    
    edge_list = []
    edge_capacities = []
    
    # We assume source is 0 and sink is hw + 1
    source = 0
    sink = h*w + 1
    
    # Add edges to source and sink
    for i in range(h):
        for j in range(w):
            node_number = coordinates_to_node(i, j, image)
            edge_list.append([source, node_number])
            edge_capacities.append(image[i,j])
            
            edge_list.append([node_number, sink])
            edge_capacities.append(255 - image[i,j])            
                
    # Iterate over the pixels and add up contributions from neigbours
    for i in range(h):
        for j in range(w):
            
            node_number = coordinates_to_node(i, j, image)
            
            # Go through neighbours
            if i>0:
                # Directly vertical neighbour
    
                # Find neigbour node number
                neighbour_node_number = coordinates_to_node(i - 1, j, image)
                
                # Add to edge list and edge capacity list
                edge_list.append([node_number, neighbour_node_number])
                edge_capacities.append(regularisation_param)
                
            if j>0:
                # Left neighbour
                # Find neigbour node number
                neighbour_node_number = coordinates_to_node(i, j - 1, image)
                
                
                # Add to edge list and edge capacity list
                edge_list.append([node_number, neighbour_node_number])
                edge_capacities.append(regularisation_param)
                


                
    graph = igraph.Graph(n = h * w + 2, edges=edge_list, directed=False, edge_attrs={'capacity': edge_capacities})
    return graph
```

```python
def denoise_image(image, regularisation_param):
    """Returns denoised image"""
    image = image.copy()
    h, w = image.shape
    
    # Construct graph using image
    graph = build_graph(image, regularisation_param)
    
    source = 0
    sink = h * w + 1
    mincut = graph.st_mincut(source, sink, capacity="capacity")

    # Get new mask generated by graphcut
    source_nodes = mincut.partition[0]

    denoised_image = np.zeros(image.shape, dtype='int')
    for node_number in source_nodes:
        if node_number!=source:
            node_coordinates = node_number_to_coordinates(node_number, image)
            denoised_image[node_coordinates] = 255
    
    return denoised_image
```

```python
denoised_yin_yang = denoise_image(noisy_image, 60)
```

```python
plt.imshow(denoised_yin_yang, cmap = 'gray')
```

# Denoising of binary images with gaussian noise

We use the following energy function for denoising of binary images:
$$
E(\mathbf{x})=\sum_{i} D_{i}\left(x_{i}\right)+\sum_{(i, j) \in \mathcal{C}} K\left|x_{i}-x_{j}\right|
$$

The data term is given as
$$
D_{i}\left(x_{i}\right)=\left\{\begin{array}{ll}
p_{i} & \text { if } x_{i}=0 \\
255-p_{i} & \text { if } x_{i}=1
\end{array}\right.
$$

where $p_i$ is the nosy image intensity.

We have a simple smoothness term with 4 connectivity assumed, and with regularisation parameter K that controls the degree of smoothness.

We construct a graph using these parameters and use mincut to find an appropriate denoised image.
